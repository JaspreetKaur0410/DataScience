```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
# LET'S EXPLORE TARGET NAMES FOR IRIS
iris.target_names
```

```python
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.tail()
```

```python
# genearate random number between 0 and 1 and if random generated number is less that .75 - its train and another .25 for test
# len(df) tells the size for uniform distribution
df['is_train'] = np.random.uniform(0,1,len(df)) <= .75
df.head(25)
```

```python
#create train and test variables
train,test = df[df['is_train'] == True], df[df['is_train'] == False]
```

```python
features = df.columns[:4]
print(features)
```

```python
#Convert species name(which is the target) ito digits
#factorize() is used to represent array in numeric form

y = pd.factorize(train['species'])[0]
y
```

```python
#Creaing a random forest classifier
clf = RandomForestClassifier(n_jobs=2, random_state=0)

#Training
clf.fit(train[features], y)
```

```python
#prediction for test data
clf.predict(test[features])
```

```python
test['features']
```

```python
predictions = iris.target_names[clf.predict(test[features])]
actual = test['species']
```

```python
result_df = pd.DataFrame({'predictions':predictions,'actual':actual})
result_df
```

```python
#creating confusion matrix
# shows that there is only 1 error in the prediction
pd.crosstab(test['species'],predictions,rownames=['Actual Species'],colnames=['Prediced Species'])
```
