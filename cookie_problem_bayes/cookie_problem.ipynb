There are 2 bags - 1st bag -> 30 Vanilla + 10 Chocolate cookie

2nd Bag -> 20 Vanilla + 20 Chocolate

A cookie is drawn at random and find that IT's VANILLA and we have to find the probability that cookie drawn at random is drawn from BOWl_A(or BOWL_B), given that it's vanilla

```python
class Cookie:
    
    def __init__(self):
        self._container = dict()
        #print(self._container)
    def Set(self,hypothesis,prob):
        self._container[hypothesis]=prob
    def Mult(self,hypothesis,prob):
        old_prob = self._container[hypothesis]
        self._container[hypothesis] = prob*old_prob
        print(self._container[hypothesis])
    def Normalize(self):
        count = 0
        for hypothesis in self._container.values():
            count = count + hypothesis
        for hypothesis,prob in self._container.items():
            self._container[hypothesis] = self._container[hypothesis]/count
    def probability(self,hypothesis):
        prob = self._container[hypothesis]
        return prob
```

Instantiate cookie object

```python
cookie = Cookie()
```

There are 2 hypothesis - Bowl_A(Bowl_A is picked up) and Bowl_B(Bowl_B is picked up) 
Set() method will set the probability for the hypothesis to 0.5

```python
cookie.Set('Bowl_A',0.5)
cookie.Set('Bowl_B',0.5)
```

Mult() will calculate prior * Likelihood
For bowl_A = (1/2)*(3/4) = 3/8 = 0.375
For bowl_B = (1/2)*(1/2) = 1/4 = 0.25

```python
cookie.Mult('Bowl_A',0.75)
cookie.Mult('Bowl_B',0.5)
```

Normalize() method will calculate the probability of Vanilla (considering 0.5 probability of both bowls)
And then calculate the posterior

```python
cookie.Normalize()
```

AND FINALY CALCUATE PROBABILITIES FOR BOWL_A AND BOWL_B, GIVEN THAT COOKIE IS VANILLA

```python
cookie.probability('Bowl_A')
```
